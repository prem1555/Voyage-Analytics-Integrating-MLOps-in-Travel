```python
# Flight Price Distribution

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

sns.histplot(flights['price'], bins=30)

plt.title('Flight Price Distribution')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='flightType', data=flights)

plt.title('Flight Types')

plt.show()
```

```python
plt.figure(figsize=(12,5))

flights['agency'].value_counts().head(10).plot(kind='bar')

plt.title('Top Flight Agencies')

plt.show()
```

```python
plt.figure(figsize=(12,5))

flights['to'].value_counts().head(10).plot(kind='bar')

plt.title('Top Destinations')

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(flights.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')

plt.show()
```

```python
sns.countplot(x='gender', data=users)

plt.title('Gender Distribution')

plt.show()
```

```python
sns.histplot(users['age'], bins=20)

plt.title('Age Distribution')

plt.show()
```
